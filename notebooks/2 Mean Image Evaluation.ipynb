# Mean Image Evaluation

We are now looking into the `MeanImageEvaluation`

```python
%load_ext autoreload
%autoreload 2
    
import numpy as np
from matplotlib import pyplot as plt
from pydatatom import GzipPickleDataset, MeanImageEvaluation

%matplotlib ipympl
```

```python
dataset = GzipPickleDataset("/tmp/AA_clock_spect_0G")
len(dataset)
```

```python
evaluation = MeanImageEvaluation()
evaluation.evaluate(dataset)
```

```python
n = evaluation.mean_image.shape[0]

fig, axes = plt.subplots(n)

for i in range(n):
    axes[i].set_title(f'{i+1}th image')
    axes[i].imshow(evaluation.mean_image[i])

plt.show()
```
